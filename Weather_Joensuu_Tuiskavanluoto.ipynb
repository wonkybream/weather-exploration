```python
import polars as pl
from pathlib import Path
import matplotlib.pyplot as plt

pl.Config.set_tbl_width_chars(140)
pl.Config.set_tbl_hide_dataframe_shape(True)
pl.Config.set_tbl_cols(-1)
pl.Config.set_tbl_rows(-1)
```

```python
# Download dataset
source = Path("data/Liperi*")

weather_df = pl.read_csv(source, has_header=True, null_values=["-"])
```

```python
# Shape and sample
print(weather_df.shape)
print(weather_df.sample(5))
```

```python
# Add "Date" column (ISO 8601 date) and translate other columns
weather_df = (
    weather_df
    .rename({
        "Havaintoasema": "Observation station",
        "Vuosi": "Year",
        "Kuukausi": "Month",
        "Päivä": "Day",
        "Aika [Paikallinen aika]": "Time [Local time]",
        "Lämpötilan keskiarvo [°C]": "Average temperature [°C]",
        "Ylin lämpötila [°C]": "Highest temperature [°C]",
        "Alin lämpötila [°C]": "Lowest temperature [°C]",
        "Keskituulen nopeus [m/s]": "Average wind speed [m/s]",
        "Tuulen suunnan keskiarvo [°]": "Average wind direction [°]",
        "Kovin puuska [m/s]": "Strongest gust [m/s]",
        "Tunnin sademäärä [mm]": "Hourly rainfall [mm]",
        "Suhteellisen kosteuden keskiarvo [%]": "Average relative humidity [%]",
    })
    .with_columns(
        Date=pl.date(pl.col("Year"), pl.col("Month"), pl.col("Day"))
    )
    .sort("Date")
)
```

```python
# Min and max dates
min_date = weather_df.select("Date").min().item()
max_date = weather_df.select("Date").max().item()

print(f"Date range for observations: {min_date} – {max_date}")
```

```python
# Print schema
for key, value in weather_df.schema.items():
    print(f"\"{key}\": {value}")
```

## Visualizing mean temperature

```python
# plot
x = weather_df.select("Date").to_series()
y = weather_df.select("Average temperature [°C]").to_series()

fig, ax = plt.subplots(figsize=(16, 4))
ax.grid(visible=True)
ax.plot(x,y)

ax.set_xlabel("Date")
ax.set_ylabel("Average temperature [°C]")

plt.show()
```

## Visualising relationship between mean temperature and wind speed

```python
# Scatter plot
x = weather_df.select("Average temperature [°C]").to_series()
y = weather_df.select("Average wind speed [m/s]").to_series()

fig, ax = plt.subplots(figsize=(16, 4))
ax.grid(visible=True)
ax.scatter(x,y, s=1)

ax.set_xlabel("Average temperature [°C]")
ax.set_ylabel("Average wind speed [m/s]")

plt.show()
```

## TMAX and TMIN per period

Calculations are done similarly as defined in [USDA – National Water and Climate Center](https://www.nrcs.usda.gov/wps/portal/wcc/home/climateSupport/wetlandsClimateTables/exampleAndDefinitions).

```python
daily_tmax_and_tmin_df = (
    weather_df
    .group_by_dynamic("Date", every="1d")
    .agg(
        tmax=pl.col("Highest temperature [°C]").max(),
        tmin=pl.col("Lowest temperature [°C]").max()
    )
    .sort("Date")
)

print(daily_tmax_and_tmin_df.shape)
print(daily_tmax_and_tmin_df.sample(10))
```

```python
yearly_snapshot_df = (
    daily_tmax_and_tmin_df
    .group_by_dynamic("Date", every="1y")
    .agg(
        tmax=pl.col("tmax").mean(),
        tmin=pl.col("tmin").mean()
    )
    .sort("Date")
)

bigger_snapshot_df = (
    yearly_snapshot_df
    .group_by_dynamic("Date", every="9y", start_by="datapoint")
    .agg(
        tmax=pl.col("tmax").mean(),
        tmin=pl.col("tmin").mean()
    )
    .sort("Date")
)

print(yearly_snapshot_df)
print(bigger_snapshot_df)
```

```python
# plot
x = yearly_snapshot_df.select("Date").to_series()
y1 = yearly_snapshot_df.select("tmax").to_series()
y2 = yearly_snapshot_df.select("tmin").to_series()

fig, ax = plt.subplots(figsize=(16, 4))
ax.grid(visible=True)
ax.plot(x, y1, "--", color="#CC6677")
ax.plot(x, y2, "--", color="#88CCEE")

ax.set_xlabel("Date")
ax.set_ylabel("TMAX and TMIN [°C]")

plt.show()
```

```python
# plot
x = bigger_snapshot_df.select("Date").to_series()
y1 = bigger_snapshot_df.select("tmax").to_series()
y2 = bigger_snapshot_df.select("tmin").to_series()

fig, ax = plt.subplots(figsize=(16, 4))
ax.grid(visible=True)
ax.plot(x, y1, "o--", color="#CC6677")
ax.plot(x, y2, "o--", color="#88CCEE")

ax.set_title("Mean of TMAX and TMIN for every 9 year period")
ax.set_xlabel("Date")
ax.set_ylabel("TMAX and TMIN [°C]")

for i, item in enumerate(zip(y1, y2)):
    ax.text(x[i], y1[i] - .12, f"{item[0]:0.2f}", fontsize=12)
    ax.text(x[i], y2[i] + .05, f"{item[1]:0.2f}", fontsize=12)

plt.show()
```
